```python
import seaborn as sns
import numpy as np
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.barplot(x='sex',y='total_bill',data=tips) #it takes mean of total bill
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) #it takes standard deviation of total bill
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue="sex")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips)
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="smoker",split=True)
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"],kde=False,bins=30)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind='scatter')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue="sex")
```

```python
tips_corr=tips.corr()
```

```python
tips_corr
```

```python
sns.heatmap(tips_corr,annot=True)
```
