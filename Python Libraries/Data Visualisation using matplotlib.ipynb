# Data Visualization

### 1. Line Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
themes=plt.style.available
print(themes)
```

```python
plt.style.use('seaborn')
```

```python
x=np.arange(10)
print(x)

y1=x**2
y2=3*x+2

print(y1)
print(y2)
```

```python
plt.plot(x,y1,color='red',label='apple',marker='o')
plt.plot(x,y2,color='green',label='kiwi',linestyle='dashed',marker='*')
plt.xlabel('time')
plt.ylabel('price')
plt.title('prices of fruits with respect to time')
plt.legend()
plt.show()
```

```python
prices=np.array([1,2,3,4])**3
print(prices)
```

```python
plt.plot(prices)
plt.show()
```

### 2. Scatter Plots

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y1,color='red',label='apple',marker='^')
plt.scatter(x,y2,color='green',label='kiwi')
plt.xlabel('time')
plt.ylabel('price')
plt.title('scatter plot - prices of fruits with respect to time')
plt.legend()
plt.show()
```

### 3. Bar Graphs

```python
plt.style.use("seaborn")
x_coordinates=np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="current year",tick_label=["gold","silver","platinum"])
plt.bar(x_coordinates+0.25,[12,10,20],width=0.5,label="next year")
plt.legend()
plt.ylim(0,40)
plt.xlim(-2,5)
plt.title("prices of metal")
plt.xlabel("metal")
plt.ylabel("price")
plt.show()
```

```python
plt.style.use("dark_background")
x_coordinates=np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="current year",tick_label=["gold","silver","platinum"],color="orange")
plt.bar(x_coordinates+0.25,[12,10,20],width=0.5,label="next year",color="blue")
plt.legend()
plt.title("prices of metal")
plt.xlabel("metal")
plt.ylabel("price")
plt.show()
```

### 4. Pie Charts

```python
plt.style.use('seaborn')
subjects=["maths","physics","chemistry","english"]
weightage=np.array([25,20,15,10])
plt.pie(weightage,labels=subjects,explode=[0,0,0.1,0],autopct='%1.1f%%',shadow=True)
plt.show()
```

### 5. Histograms

```python
xsn=np.random.randn(100)
sigma=8
u=70
x=np.round(xsn*sigma+u)
x2=np.round(xsn*5+40)
print(x,x2)
```

```python

plt.hist(x2,alpha=0.8,label="maths")
plt.hist(x,alpha=0.9,label="physics")
plt.xlabel('marks range')
plt.ylabel('prob/freqCount of students')
plt.title('histogram')
plt.legend()
plt.show()
```

```python
u=5
sigma=1

vals=u+sigma*np.random.randn(100)

print(vals.shape)

plt.hist(vals,10) #10 is number of buckets
plt.show()
```

```python
x=vals
y=np.zeros(x.shape)

plt.scatter(x,y)
plt.show()
```

### Multi Variate Normal Distribution

```python
#bi variate
u=np.array([0.0,0.0])
cov=np.array([[1.0,0.0],[0.0,1.0]])

dist=np.random.multivariate_normal(u,cov,500)
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```

```python
u=np.array([0.0,0.0])
cov=np.array([[1.0,0.8],[0.8,1.0]])

dist=np.random.multivariate_normal(u,cov,500)
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```

```python
u=np.array([0.0,0.0])
cov=np.array([[1.0,0.8],[0.8,1.0]])
u2=np.array([5.0,6.0])
cov2=np.array([[1.2,0.3],[0.2,1.1]])

dist=np.random.multivariate_normal(u,cov,500)
dist2=np.random.multivariate_normal(u2,cov2,500)


plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
